```julia
using POMDPs
using Perseus
using POMDPModels
using POMDPSimulators: RolloutSimulator, simulate
```

```julia
pomdp = BabyPOMDP() # initialize POMDP

solver = PerseusSolver(pomdp, max_iterations=10000, depth=6, branchingfactor=0, verbose=true)

# run the solver
policy1 = solve(solver, pomdp)   # policy is of type AlphaVectorPolicy
```

```julia
m = BabyPOMDP(-5, -10, 0.1, 0.8, 0.1, 0.99)

solver = PerseusSolver(m, max_iterations=10000, depth=10, branchingfactor=0, improvement_tolerance=1e-15, verbose=true)
policy = solve(solver, m)

println(policy)

μ = zeros(10000)

for i in 1:10000
    print(i, "\r")
    rollout_sim = RolloutSimulator(max_steps=1000)
    μ[i] += simulate(rollout_sim, m, policy)
end

println(sum(μ)/10000)
```
