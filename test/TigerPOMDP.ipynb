```julia
using POMDPs
using Perseus
using POMDPModels
using POMDPModelTools
#using POMDPSimulators: RolloutSimulator, simulate
```

```julia
pomdp = TigerPOMDP() # initialize POMDP

solver = PerseusSolver(pomdp, max_iterations=10000, depth=6, branchingfactor=0, verbose=true)

# run the solver
policy1 = solve(solver, pomdp)   # policy is of type AlphaVectorPolicy
```

### benchmark

```julia
using BenchmarkTools
```

```julia
POMDPModelTools.ordered_actions(pomdp::TigerPOMDP) = [0, 1, 2]
POMDPModelTools.ordered_states(pomdp::TigerPOMDP) = [false, true]
POMDPModelTools.ordered_observations(pomdp::TigerPOMDP) = [false, true]
```

```julia
pomdp = TigerPOMDP()
@benchmark solver = PerseusSolver(pomdp, max_iterations=10000, depth=6, branchingfactor=15)
```

```julia
pomdp = TigerPOMDP()
solver = PerseusSolver(pomdp, max_iterations=10000, depth=6, branchingfactor=0)

# run the solver
@benchmark policy1 = solve(solver, pomdp)   # policy is of type AlphaVectorPolicy
```
